```python
import matplotlib.pyplot as plt
import distributions as dist
```

```python
N = 100
gens = 10
start_p = 0.5
x_axis = [i for i in range(gens+1)]

binom = dist.binomial(N,start_p)
for i in range(10):
    binom.p = start_p
    freqs = [start_p]
    for gen in range(gens):
        n_allele1 = binom.sample()[0] 
        freq1 = float(n_allele1) / float(N)
        freqs.append(freq1)
        binom.p = freq1
    plt.plot(x_axis,freqs)

plt.show()
    

    
```

```python


def sim_wf(n, start_freq = 0.5, ngen = 1000):
    proc_freqs=[start_freq]
    binom = dist.binomial(n,start_freq)
    for gen in range(ngen):
        freq_next = binom.sample()[0] / n
        binom.p = freq_next
        proc_freqs.append(freq_next)
    return proc_freqs
        
n = 1000
start_p = 0.5
#sel = 0.000001
ngen = 1000
nrep = 40
for i in range(nrep):
    walk = sim_wf(n,start_p,ngen)
    plt.plot(range(len(walk)),walk)
    plt.ylim([-.1, 1.1])
plt.show()
```

```python
nrep = 90

tslices = [[],[],[]]
for i in range(nrep):
    walk = sim_wf(n,start_p,ngen)
    tslices[0].append(walk[int(ngen/4)-1])
    tslices[1].append(walk[int(ngen/2)-1])
    tslices[2].append(walk[int(ngen)-1])
    plt.plot(range(len(walk)),walk)
    plt.ylim([-.1, 1.1])
plt.show()
```

```python
import math

def calc_mean(vals):
    tally = 0.0
    for i in vals:
        tally += i
    mean = tally / float(len(vals))
    return mean

def calc_var(vals):
    tally = 0.0
    mean = calc_mean(vals)
    for i in vals:
        tally += ( i - mean ) ** 2.0
    
    var = tally / float(len(vals))
    return var
```
