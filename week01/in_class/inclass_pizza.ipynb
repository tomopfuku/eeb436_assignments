```python
fl = open("pizza.txt", "r")
```

```python
print(fl)
```

```python
lines = fl.readlines()
print(lines[0])
```

```python
pizza_cities = {}
for line in lines[1:]:
    spls = line.strip().split(",") 
    city = spls[1]
    try:
        pizza = float(spls[3])
    except:
        continue
    pizza_cities[city] = pizza
```

```python
import matplotlib.pyplot as plt
```

```python
cities = list( pizza_cities.keys() )[0:10]
pizzas = list( pizza_cities.values() )[0:10]
```

```python
plt.bar(cities,pizzas)
plt.xticks(rotation=30)
```
